```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.boxofficemojo.com/year/world/?ref_=bo_nb_rs_tab'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')
```

```python
table = soup.find('tr')
```

```python
table
```

```python
table_name = table.find_all('th')
```

```python
table_name
```

```python
heading_title = [header.text.strip() for header in table_name]
```

```python
heading_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = heading_title)
```

```python
df
```

```python
table_data = soup.find_all('tr')
```

```python
for row in table_data:
    row_data = row.find_all('td')
    unique_row_data = [data.text.strip() for data in row_data]
    print(unique_row_data)
```

```python
for row in table_data:
    row_data = row.find_all('td')
    unique_row_data = [data.text.strip() for data in row_data]

    if len(unique_row_data) == len(df.columns):
        length = len(df)
        df.loc[length] = unique_row_data
    else:
        print(f"skipping row due to mismatch: {unique_row_data}")
```
